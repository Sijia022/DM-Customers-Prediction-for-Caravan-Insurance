## 1.3 Split the data for  training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
# reshuffle the data
caravan = caravan.sample(frac=1, random_state=42)

X = caravan.iloc[:, :-1]
y = caravan.iloc[:, -1]
```

```python
# split the test data (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)
```

```python
print(f"number of training data is {len(X_train)}")
print(f"number of testing data is {len(X_test)}")
```

## 1.4 Resample only the training data

```python
from sklearn.utils import resample
```

```python
train_data = pd.concat([X_train, y_train], axis=1)

# select of positive and negative classes
train_data_positive = train_data.loc[caravan['CARAVAN POLICY'] == 1]
train_data_negative = train_data.loc[caravan['CARAVAN POLICY'] == 0]

num_sample_each = int(len(train_data)/2)

# undersample negative class
train_data_negative_undersample = resample(train_data_negative, replace=False,n_samples=num_sample_each)

# oversample positive class
train_data_positive_oversample = resample(train_data_positive, replace=True,n_samples=num_sample_each)

# combine two sampled classes
train_data_over_under_sample = pd.concat([train_data_negative_undersample, train_data_positive_oversample])

# reshuffle the training data
train_data_over_under_sample = train_data_over_under_sample.sample(frac=1, random_state=42)
```

```python
# look at the distribution of the class y again
y = train_data_over_under_sample['CARAVAN POLICY']

# create a histogram to show the difference
bin_edges = [-0.5, 0.5, 1.5]

plt.figure(figsize=(6, 4))
plt.hist(y, bins=bin_edges, rwidth=0.8)

plt.xticks(range(2))
plt.xlabel('Class')
plt.ylabel('Numbers')
plt.title('Distribution of Two Classes')
    
plt.show()
```

## 1.4 Encode two normial data types using target encoding

```python
train_data = train_data_over_under_sample

# target encoding in the training set
Customer_Subtype_mean_encoding = train_data.groupby('Customer Subtype')['CARAVAN POLICY'].mean()
train_data['Customer Subtype'] = train_data['Customer Subtype'].map(Customer_Subtype_mean_encoding)

Customer_main_type_mean_encoding = train_data.groupby('Customer main type')['CARAVAN POLICY'].mean()
train_data['Customer main type'] = train_data['Customer main type'].map(Customer_main_type_mean_encoding)

X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]
```

```python
# apply the same encoding to test set (no leakage)
X_test['Customer Subtype'] = X_test['Customer Subtype'].map(Customer_Subtype_mean_encoding)
X_test['Customer main type'] = X_test['Customer main type'].map(Customer_main_type_mean_encoding)
```

## 1.5 Feature selection for description of target group

### 1.5.1 Filter

```python
# 1. Chi-Square Test for Independence
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X = X_train
y = y_train

selector = SelectKBest(chi2, k=10)
selector.fit_transform(X, y)


# get the scores for each feature
feature_scores = selector.scores_

# create a DataFrame to store the scores along with feature names
feature_scores_df = pd.DataFrame({'Feature': X.columns, 'Feature_Score': feature_scores})

# sort the DataFrame by scores in descending order
feature_scores_df = feature_scores_df.sort_values(by='Feature_Score', ascending=False)

# visualize the top 15 features
top_15_features = feature_scores_df.head(15)


plt.figure(figsize=(8, 4))
plt.barh(top_15_features['Feature'], top_15_features['Feature_Score'])
plt.xlabel('Chi-Squared Score')
plt.title('Top 15 Features by Chi-Squared Score')
plt.gca().invert_yaxis()
plt.show()

#store top features for compare among different models later
features_comparison = top_15_features['Feature'].values
```

```python
# 2. Mutual information
from sklearn.feature_selection import mutual_info_classif

selector = SelectKBest(mutual_info_classif, k=10)
selector.fit_transform(X, y)


# calculate mutual information scores for each feature
mi_scores = mutual_info_classif(X, y)

# create a DataFrame to store the scores along with feature names
mi_scores_df = pd.DataFrame({'Feature': X.columns, 'MI_Score': mi_scores})

# sort the DataFrame by MI_Score in descending order
mi_scores_df = mi_scores_df.sort_values(by='MI_Score', ascending=False)

# visualize the top 15 features
top_15_features = mi_scores_df.head(15)


plt.figure(figsize=(8, 4))
plt.barh(top_15_features['Feature'], top_15_features['MI_Score'])
plt.xlabel('Mutual Information Score')
plt.ylabel('Feature')
plt.title('Top 15 Features by Mutual Information Score')
plt.gca().invert_yaxis()  # Invert the y-axis to show the most important feature at the top
plt.show()

#store top features for compare among different models later
features_comparison = np.append(features_comparison,top_15_features['Feature'].values)
```

## Model 1
### 2.1 chi2 for feature selection + Logistic regression model

```python
# combine feature selection in training classification models
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

X = X_train
y = y_train

classifier = LogisticRegression(max_iter=1000)
chi2_logist_pipeline = make_pipeline(SelectKBest(chi2, k=10), classifier)

chi2_logist_pipeline.fit(X, y)
```

```python
# evaluate the result with validation
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import confusion_matrix
from tabulate import tabulate

MyClassifier = chi2_logist_pipeline 

# 1-get vaildation scores
scores = cross_val_score(MyClassifier, X, y, cv=10)
mean_accuracy = scores.mean()
print(f"Mean Accuracy: {mean_accuracy:.2f}")

# 2-get confusion matrix
y_val_pred = cross_val_predict(MyClassifier, X, y, cv=10)
conf_mat = confusion_matrix(y, y_val_pred)

# extract values from the confusion matrix
TN, FP, FN, TP = conf_mat.ravel() 

recall = TP / (TP + FN)
precision = TP / (TP + FP)
F1_Score = 2 * (precision * recall) / (precision + recall)
print("Recall:", recall)
print("Precision:", precision)
print("F1_Score:", F1_Score)

print("Confusion Matrix:")
columns = ['Pos', 'Neg']
index = ['Pos', 'Neg']
# convert the matrix into a DataFrame with column names and row names
df = pd.DataFrame([[TP,FN],[FP,TN]], columns=columns, index=index)
table = tabulate(df, headers='keys', tablefmt='grid')
print(table)
```

```python
# make predictions on the test set
from sklearn.metrics import accuracy_score

y_pred = MyClassifier.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
print("Test accuracy:", test_accuracy)

conf_mat = confusion_matrix(y_test, y_pred)

# extract values from the confusion matrix
TN, FP, FN, TP = conf_mat.ravel() 

recall = TP / (TP + FN)
precision = TP / (TP + FP)
F1_Score = 2 * (precision * recall) / (precision + recall)
print("Recall:", recall)
print("Precision:", precision)
print("F1_Score:", F1_Score)

print("Confusion Matrix:")
columns = ['Pos', 'Neg']
index = ['Pos', 'Neg']
# convert the matrix into a DataFrame with column names and row names
df = pd.DataFrame([[TP,FN],[FP,TN]], columns=columns, index=index)
table = tabulate(df, headers='keys', tablefmt='grid')
print(table)

# collect test results for model comparison later
test_accuracy_all = np.array(test_accuracy)
recall_all = np.array(recall)
precision_all = np.array(precision)
F1_score_all = np.array(F1_Score)
```
